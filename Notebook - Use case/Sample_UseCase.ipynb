```python
def evol( nqubits, nparameters, gamma=10, output="coherent_kernel_result.csv"):
    start = time.time()
    pop, pareto, logbook = gsvm.coherent_gsvm(nqubits=nqubits, depth=6, nparameters=nparameters,
                                     weights=[-1.0,-1000.0],
                                     mu=50,lambda_=10, ngen=10000, gamma=gamma)
    print(f'Simulation finished after {time.time()-start} seconds')
    print(f'\nGenetic-algorithm output ({output})')
    #print('generation, individual, gates/qubit, accuracy')
    print('---------------------------------------------')
    with open(output, "w") as f:
        for ide, ind in enumerate(pareto):
            genes=''.join(str(i) for i in list(ind))
            gates, acc = ind.fitness.values
            line = f'{ide},"{genes}",{gates},{acc}'
            f.write(line)
            f.write('\n')
            print(line)

nqubits = 3
nparameters = 1
gamma=10
evol(nqubits=nqubits, nparameters=nparameters, gamma=gamma)
```

```python
# import encoding2 as e2
# import random

iot_result = pd.read_csv('coherent_kernel_result.csv',header=None)

def ordenar_salidas_pareto(dataframe):
    dataframe.columns=['ind','circ','gates','acc']
    dataframe.sort_values(['acc','gates'], ascending=[False,False],inplace=True)
    dataframe.reset_index(inplace=True)
    dataframe.pop('index')
    return dataframe

iot_salidas = ordenar_salidas_pareto(iot_result)
iot_salidas # Queremos ver el primer circuito del dataframe yya ordenado por accuracy (mejor ind)
```

```python
import random
from fitness import CoherentFitness
from encoding import CircuitConversor

arange = np.linspace(-1, 1, 100)

cc = CircuitConversor(nqubits=nqubits, nparameters=nparameters)

pop = iot_salidas['circ'][4]

params = np.random.uniform(-0.8, 0.8, size=(1, 1000))
states, costs = cc(pop, params)

diffs = []
kernel_values = []
true_kernel_values = []
for i in range(1000):
    idx1 = random.randint(0, len(arange)-1)
    idx2 = random.randint(0, len(arange)-1)
    state1 = states[idx1]
    state2 = states[idx2]
    diff = params[0, idx1] - params[0, idx2]
    kernel_value = np.abs(np.dot(state1.conj(), state2))**2
    diffs.append(diff)
    kernel_values.append(kernel_value)
    true_kernel_values.append(CoherentFitness.gaussian_kernel(params[0:, idx1:idx1+1], params[0:, idx2:idx2+1], gamma=gamma))
    
plt.scatter(diffs, kernel_values, label='Kernel values')
plt.scatter(diffs, true_kernel_values, label='True kernel values')
plt.legend()
plt.show()
```
